### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
# Create a 2x2 subplot layout
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart (Top-Left)
x = np.linspace(0, 10, 100)
y = x ** 2
axes[0, 0].plot(x, y, color='b', linewidth=2)
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values')
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# Bar Chart (Top-Right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter Plot (Bottom-Left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
for i in range(len(x_scatter)):
    axes[1, 0].text(x_scatter[i], y_scatter[i], f'({x_scatter[i]}, {y_scatter[i]})', fontsize=10, ha='right', va='bottom')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X Values')
axes[1, 0].set_ylabel('Y Values')
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie Chart (Bottom-Right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)
axes[1, 1].set_title('Pie Chart')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
